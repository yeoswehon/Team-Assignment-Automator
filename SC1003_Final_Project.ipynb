##**Merge Step in the Merge Sort Algorithm**

This section of the code implements the merge step of the famously known algorithm, Merge Sort. This merge function is responsible for merging two sorted subarrays into a larger array, ensuring that this new larger array is sorted after merging.

How the algorithm works:

**Divide the array by half**

1. The array is divided into two halves, using the `mid` index, with the first half of the array spanning from the `left_idx` to the `mid` index, and the second half of the array spaning from the `mid + 1` index to the `right_idx`.

2. The `first_half` variable and the `second_half` variable store the lengths of the two subarray respectively.

**Creating Temporary Arrays**

3. Two arrays with sizes `first_half` and `second_half` are initialised, with each element being `0`.

4. Using a for-loop, the loop iterates to copy the elements of the original array into the `L` and `R` arrays respectively.

**Merging the Two Sorted Halves**

5. Three trackers `i`, `j`, `k` are initialised to track the position of the `L` array, `R` array and the original array respectively.

6. The `while` loop compares elements in `L` and `R`, slowly placing the smaller of the two elements into the original array. This process will terminate only until one of the trackers `i` or `j` is equal to the length of its own subarray. Each time an element is added, either `i` or `j` increments by one, and `k` will be incremented as well.

**Handling Remaining Elements**

7. In the final step, the two remaining `while` loops is to handle any remaining elements in `L` or `R` by adding the elements back into the original array, thereby ensuring that elements spanning from `left_idx` to `right_idx` is sorted.

```python
# Merge Step of Merge Sort Algorithm
# Time Complexity:
# Best Case: O(n)
# Worst Case: O(n)

# Function sorts array from left index to right index
def merge(arr, left_idx, mid, right_idx):

  # Split them into two halves
  first_half = mid - left_idx + 1
  second_half = right_idx - mid

  # Create two temporary arrays
  L = [0] * first_half
  R = [0] * second_half

  # Store the values into the temporary arrays
  for i in range(0, first_half):
    L[i] = arr[i + left_idx]

  for i in range(0, second_half):
    R[i] = arr[i + mid + 1] # Starting from index mid + 1

  i = 0 # Index tracking array L
  j = 0 # Index tracking array R
  k = left_idx # Index tracking merged array

  while i < first_half and j < second_half:
    if L[i] < R[j]:
      arr[k] = L[i]
      i += 1
    else:
      arr[k] = R[j]
      j += 1
    k += 1

  # Handle the leftover unsorted values
  while i < first_half:
    arr[k] = L[i]
    i += 1
    k += 1

  while j < second_half:
    arr[k] = R[j]
    j += 1
    k += 1
```

##**Insertion Sort Algorithm**
This section of the code implements the Insertion Sort algorithm, which is particularly effective for sorting small or nearly sorted data sets. The purpose of the function is to sort a segment from `left_idx` to `right_idx`.

How the algorithm works:

**Initial Step**
1. The algorithm starts with the first element at the specified range of `left_idx + 1` to the `right_idx`, with the first element being at index `left_idx + 1`.
2. Subsequently, it looks at the next `right_idx - left_idx - 1` elements one by one.

**Inserting the Element**
3. Looking at the current element, it "compares" itself to the element on the left to see if it is smaller.
4. It is worthy to note that everytime we perform this comparison, all elements to the left of it are already in sorted order, excluding the current element we are looking at.
5. If the current element is smaller than the one on the left, swap their positions. This process continues until the element on the left is smaller than the current element we are looking at.

**Terminating the Program**

6. The program will continue until the elements from the `left_idx` to `right_idx` is sorted. Only then will the algorithm be completed.

```python
# Insertion Sort Algorithm
# Time Complexity:
# Best case: O(n) for small inputs
# Worst case: O(n^2) for large amount of input

# Function sorts array from left index to right index
def insertion_sort(arr, left_idx, right_idx):
  for i in range(left_idx + 1, right_idx + 1):
    j = i # index of current element

    # while current index is greater than left bound, and current element is smaller than previous element
    while j > left_idx and arr[j] < arr[j-1]:
      arr[j], arr[j-1] = arr[j-1], arr[j] # Swap them
      j -= 1 # Decrement j counter by 1
```

##**TimSort Algorithm**
This section of the code implements the TimSort algorithm, which is a hybrid sorting algortihm that encompasses both the efficiency of Insertion Sort in sorting small arrays and the efficiency of Merge Sort's merge function in merging large, sorted arrays. Timsort has a worst-case time complexity of `O(nlogn)` and a best-case time complexity of `O(n)`. Before we get into the algorithm, we have to understand fundamentally, what a "min run" is.


A run is a continguous subarray within an array that is sorted or will be sorted before merging. Therefore, a min run is a threshold that defines the minimum size of a subarray that TimSort will process before merging. More formally, min run is the minimum length threshold for a run in the TimSort algorithm.


In TimSort, each subarray or run, with a size smaller than or equal to the min run will be sorted individually using insertion sort before merging through the merge function. Therefore, knowing the optimal value for the min run will lead to the most efficient sorting time. But, what exactly is this value? Ideally, TimSort wants the value of `(number of elements in array) / (min run)` to equal to a power of 2 or be as close to it as possible so that when the merge function is called, it obtains a "balanced" merge. This is corroborated by the fact that unbalanced merges introduce extra comparisons and worsen the performance (Medium, 2018).


After each run is sorted, the merge functions operates in a bottom-up fashion, where it merges smaller runs together, and then merges these new larger runs together and this process continues until one sorted array remains.


How the algorithm works:

**Calculating the Optimal Min Run**
1. The helper function `calc_min_run` calculates the optimal min run size based on the length of the array, `n`.
2. Using an arbitrary threshold of `64`, if the value of `n` exceeds the threshold `(n > threshold)`, it iteratively halves the value of `n` until it falls below the threshold, while keeping track of any "discarded" bit. This discarded bit, `r`, is a flag for whether `n` is an even power of two. If `n` is not a power of two, `r` will be `1`. Else, `r` is `0`.
3. After reducing `n` down to a manageable size, the function returns `n + r`, which is the optimal min run size for efficient sorting and merging in the future.

**Sorting Runs initially Using Insertion Sort**
4. Insertion Sort is efficient for small inputs. With our min run being small enough, it allows TimSort to quickly sort all the runs.
5. The algorithm first sorts subarrays of size min run using Insertion Sort, producing several sorted subarrays.

**Merging Sorted Runs using the Merge Function**
6. The current block size to be merged together, set as variable `current_block_size`, is initialised at value min run and doubles after each merging pass, thereby allowing larger sorted segments to be combined progressively.

7. Each pair of adjacent runs will be sorted into a larger subarray through the merge function. The variables `left_idx`, `mid` and `right_idx` define the boundaries of the subarray of size `2 * current_block_size`.

8. `left_idx` marks the start of the first segment, `mid` the end of the first segment and `right_idx` the end of the second segment. These parameters are pivotal to pass into the merge function for the sorting to occur.

9. Once the `current_block_size` doubles until it is greater than or equal to n, the while loop terminates and the function returns the sorted array.

```python
# TimSort Algorithm (Combining Insertion Sort and Merge Sort together)
# Time Complexity:
# Best case: O(n)
# Worst case: O(nlogn)

# min run is a parameter that defines the minimum size of a run that the algorithm will process
# it ensures all runs are at least min_run elements long
# adjusting the min_run based on input size ensures the sorting algorithm adapts to various data sizes

# the aim of this function is to minimise the number of runs by ensuring runs are sufficiently large
def calc_min_run(n):
  threshold = 64 # upper threshold for run size
  r = 0 # flag to keep track of any "discarded" bits which only occur if n is not a perfect power of two
  while n >= threshold:
    r |= n & 1 # Capture the least significant set bit
    n //= 2  # divide n by 2 (floor)
  minrun = n + r
  return minrun

def tim_sort(arr):
  # Run is like the "block size"
  # Optimal min_run by research is between 32 and 64
  # This is because runs of these sizes are typically cache-friendly, enhancing performance by improving cache utilisation
  n = len(arr)
  minrun = calc_min_run(n)

  # Sort the subarrays of blocks with size minrun
  for left_idx in range(0, n, minrun):
    right_idx = min(n-1, left_idx + minrun - 1) # right bound of current block
    insertion_sort(arr, left_idx, right_idx) # sort from left_idx to right_idx, which is the current block

  current_block_size = minrun

  # merge sorted blocks in a bottom-up fashion
  while current_block_size < n:
    # iterate over the array in steps of 2 * current_block_size
    for left_idx in range(0, n, 2 * current_block_size):
      mid = min(n - 1, left_idx + current_block_size - 1) # right index of first run
      right_idx = min(n - 1, left_idx + 2 * current_block_size - 1) # right index of second run

      if mid < right_idx:
        # merge the two sorted runs from left_idx to mid and mid+1 to right_idx
        merge(arr, left_idx, mid, right_idx)

    current_block_size *= 2 # multiply block size by 2
  return arr
```

##**CGPA Quartile Assignment**
The assign_cgpa function takes in the parameter student_list, which is an array of students' information in that tutorial group, and categorises students into quartiles `(q1, q2, q3, q4)` based on their cgpa, assigning every student a category based on their cgpa `(q1, q2, q3, q4)` that reflects their relative standing in their respective tutorial groups.

**Forming the CGPA List**
1. The function creates a list, `student_cgpa_list`, containing all `50` students' information in a particular tutorial group, sorted in ascending order using the TimSort algorithm.

**Calculating Quartile Boundaries**
2. The variables low, medium and high represent the cgpa scores at `q1`, `q2`, and `q3` respectively.
3. They are calculated as `student_cgpa_list[n // 4]`, `student_cgpa_list[2 * n // 4]` and `student_cgpa_list[3 * n // 4]` respectively.

**Assigning CGPA Categories**
4. The function compares their CGPA to the quartile thresholds and assigns their category accordingly. The quartile thresholds below represent their quartile range.

`q1` : Top 75-100% (below the 1st quartile)

`q2` : Top 50-75% (between the 1st and 2nd quartiles)

`q3` : Top 25-50% (between the 2nd and 3rd quartiles)

`q4` : Top 0-25% (above the 3rd quartile)

```python
def assign_cgpa(student_list):
  # form a list of the students' cgpa in the current tutorial group
  # sort the cgpa scores in ascending order
  student_cgpa_list = tim_sort([student['cgpa'] for student in student_list])
  n = 50 # Each tutorial group has 50 students
  low = student_cgpa_list[n // 4] # quartile 1
  medium = student_cgpa_list[2 * n // 4] # quartile 2
  high = student_cgpa_list[3 * n // 4] # quartile 3
  for student in student_list:
    student_cgpa = student['cgpa']
    if student_cgpa <= low:
      student['cgpa_category'] = 'q1' # Top 75-100%
    elif student_cgpa <= medium:
      student['cgpa_category'] = 'q2'
    elif student_cgpa <= high:
      student['cgpa_category'] = 'q3'
    else:
      student['cgpa_category'] = 'q4' # Top 0-25%
```

##**Team Initialisation with Detailed Member Attributes**
The make_team function takes in a parameter, number_of_teams, which is the specified number of teams that the function will initialise. For each team, a dictionary with the variable current_team will be created to store the team details such as


`team` : The team number, starting from `1`

`team_members` : An empty list to store the individual student information (`student_info`)

`school_count` : A dictionary where the school name is a key, and its value is the number of team members from that school

`gender_count` : A dictionary where the gender is the key, and its value is the gender count respectively

`cgpa_count` : A dictionary with CGPA with quartile categories as its keys (`q1`, `q2`, `q3`, `q4`), and the values denoting the number of students in that quartile category

`current_size` : Tracks the current number of members on the team, intiially set to `0`


Every single initialised `current_team` dictrionary is added to the teams array using `teams.append(current_team)`, and this teams list is the collection of all the teams in a single tutorial group.

```python
def make_team(number_of_teams):
    # Parameters:
    # - number_of_teams (int): The total number of teams to create.
    # - team_size (int): The maximum number of members per team.
    # - schools (list): List of all possible schools.
    # - genders (list): List of all possible genders.
    schools = [
    'CCDS', 'EEE', 'CoB (NBS)', 'SoH', 'WKW SCI', 'CoE', 'MAE',
    'SPMS', 'SBS', 'SSS', 'ASE', 'NIE', 'ADM', 'CCEB', 'MSE',
    'LKCMedicine', 'CEE', 'HASS'
    ]
    genders = ['Male', 'Female']
    teams = []
    for i in range(number_of_teams):
        current_team = {
            'team' : i + 1, # 0 based indexing so add 1
            'team_members' : [], # for storage of student_info
            'school_count' : {school: 0 for school in schools}, # Initialise school_count to 0
            'gender_count' : {gender: 0 for gender in genders}, # Initialise gender_count to 0
            'cgpa_count' : {'q1': 0, 'q2': 0, 'q3': 0, 'q4': 0},
            'current_size' : 0,
        }
        teams.append(current_team)
    return teams
```

```python
def update_data(curr_team, student, student_gender, student_cgpa, student_school):
  '''
  Assign a student to the team and update the team's statistics accordingly

  Parameters:
  curr_team (dictionary) : the current team's data structure
  student (dictionary) : the student's information (gender, school, student_id, etc)
  student_gender (string) : male or female
  student_cgpa (string) : the cgpa quartile of the student ('q1', 'q2', 'q3', 'q4')
  student_school (string): the school the student belongs to

'''
  curr_team['current_size'] += 1 # increment team size by 1
  student['team_assigned'] = True # flag having a team
  curr_team['team_members'].append(student) # add student_info to the current team

  curr_team['gender_count'][student_gender] = curr_team['gender_count'].get(student_gender, 0) + 1 # Increment gender count by 1
  curr_team['cgpa_count'][student_cgpa] = curr_team['cgpa_count'].get(student_cgpa, 0) + 1 # Increment cgpa count by 1
  curr_team['school_count'][student_school] = curr_team['school_count'].get(student_school, 0) + 1 # Increment school count by 1
```

##**Main Algorithm**
The code below is an algorithm that assigns students to teams while balancing criteria such as gender, CGPA distribution and school affiliation.

**Initial Setup**
1. `total_students` : Sets the total number of students, as per required in the project details
2. `number_of_teams` : Calculated by dividing the total number of students by the size of each team
3. `majority_gender` : Threshold for a "majority" in terms of gender representation within a team
4. `majority_cgpa` : Maximum count for each CGPA quartile in a team
5. `majority_school` : To limit the number of students from the same school to at most `2`

**Answer Dictionary**

`answer = {}` : initialises an empty dictionary to store all teams for each tutorial group, with the key being the tutorial group number and the value being a list that contains all the teams for that particular tutorial group number


**Team Initialisation and CGPA Quartile Assignment**
1. `for tutorial_group_number, student_list in tutorial_groups.items()` : Iterate over each tutorial group and its list of students
2. Assign the CGPA quartile to students in the particular tutorial group by invoking the `assign_cgpa` function
3. Initialise empty teams for the tutorial groups by invoking the `make_team` function
4. `current_student_count` : Tracks the count of students processed within the group

**Extracting Student Detail**
1. `for student in student_list` : Iterates over each student in the tutorial group
2. Extract student details such as their gender, cgpa quartile and school
3. Variable `opposite_gender` is set to the gender opposite of the current student

**Primary Assignment of Student to Team (Primary Criteria)**
1. `for curr_team in teams` : Iterate over each team
2. Checks for the following conditions:

  (a) Fewer students of the current gender than the opposite gender

  (b) No student in the CGPA quartile is assigned yet

  (c) No student from the same school

  (d) Available space `(current_team_size < team_size)`
3. If conditions are met, assign student to the current team by executing the `update_data` function

**Fallback Assignment of Student to Team (Secondary Criteria)**

If the primary criteria is not met, it means the student is still unassigned to a team. In that case, we shall proceed with an alternative criteria, applying each criterion in the following order to assign the student. Once a criterion is met, the student will be assigned to the team by executing the update_data function.

The criteria are applied in the following order of priority:

1. Assign student to team if adding this student does not cause the team's gender balance, cgpa distribution and school distribution to exceed the `majority_gender`, `majority_cgpa` and `majority_school` limits respectively

2. Assign student based on `majority_cgpa` limit

3. Assign student based on `majority_gender` limit

4. Assign student based to any team with available space if none of the above criteria are met `(current_team_size < team_size)`


**Edge Cases (Additional Question Requirement)**

If the `team_size` is not a multiple of 5, it is impossible to assign the students evenly across all the teams.

To address this, if some students remain unassigned after all the criteria are applied, the remaining unassigned students will be distributed across teams in a round-robin fashion.

To achieve this, the line `if (current_student_count - 1) % number_of_teams == i:` performs the round-robin assignment by distributing the teams in ascending order (i.e from `team 1` to `team 2`, etc).

**Storing Teams in Answer Dictionary**

`if tutorial_group_number not in answer` : Initialises an empty list in the `answer` dictionary if it does not exist
`answer[tutorial_group_number].append(curr_team)` : Appends each finalised team to the `answer` dictionary under its respective key, the `tutorial_group_number`. Here, the value is a list of teams assigned to that tutorial group, which will eventually be used later on for output purposes


```python
total_students = 50 # from the problem statement
number_of_teams = (total_students // team_size)
majority_gender = team_size // 2 # how we define majority of gender to be
majority_cgpa = team_size // 4 # how we define majority of cgpa to be
majority_school = 1 # how we define majority of school to be

answer = {} # key : tutorial group number, value: array containing all the teams for the tutorial group number

for tutorial_group_number, student_list in tutorial_groups.items():
  assign_cgpa(student_list) # assign their respective quartiles (q1 - q4)
  teams = make_team(number_of_teams) # initialise empty teams
  current_student_count = 0

  for student in student_list:
    current_student_count += 1 # increment current student count by 1
    student_gender = student['gender'] # current student gender
    student_cgpa = student['cgpa_category'] # current student cgpa
    student_school = student['school'] # current student school

    opposite_gender = 'Female' if student_gender == 'Male' else 'Male'

    for curr_team in teams:
      # if (# of current gender <= # of opposite_gender) and team does not have this quartile and school and does not exceed the highest possible group size
      if (curr_team['gender_count'].get(student_gender, 0) <= curr_team['gender_count'].get(opposite_gender, 0) and # Assign uniquely if this unique combination does not exist in team
        curr_team['cgpa_count'].get(student_cgpa, 0) == 0 and
        curr_team['school_count'].get(student_school, 0) == 0 and
        curr_team['current_size'] < team_size):
        update_data(curr_team, student, student_gender, student_cgpa, student_school)
        break # break out of the loop

    if not student['team_assigned']: # current student still does not have a team
      for curr_team in teams:
        # if current_team's gender count is less than the majority
        # current cgpa_count (quartile) is less than the majority
        # and we only allow maximum of two of the same school
        # and the current team size does not exceed the highest possible group size
        if (curr_team['gender_count'].get(student_gender, 0) < majority_gender and
        curr_team['cgpa_count'].get(student_cgpa, 0) < majority_cgpa and
        curr_team['school_count'].get(student_school, 0) <= majority_school and
        curr_team['current_size'] < team_size):
          update_data(curr_team, student, student_gender, student_cgpa, student_school)
          break # break out of the loop

    if not student['team_assigned']: # current student still does not have a team
      for curr_team in teams:
        # assign based on majority_cgpa count and make sure it does not exceed the highest possible group size
        if (curr_team['cgpa_count'].get(student_cgpa, 0) < majority_cgpa and
        curr_team['current_size'] < team_size):
          update_data(curr_team, student, student_gender, student_cgpa, student_school)
          break # break out of the loop

    if not student['team_assigned']: # current student still does not have a team
      for curr_team in teams:
        # assign based on majority_gender count and make sure it does not exceed the highest possible group size
        if (curr_team['gender_count'].get(student_gender, 0) < majority_gender and
        curr_team['current_size'] < team_size):
          update_data(curr_team, student, student_gender, student_cgpa, student_school)
          break # break out of the loop

    if not student['team_assigned']: # current student still does not have a team
      for curr_team in teams:
        # assign based on school count and make sure it does not exceed the highest possible group size
        if (curr_team['school_count'].get(student_school, 0) <= majority_school and
        curr_team['current_size'] < team_size):
          update_data(curr_team, student, student_gender, student_cgpa, student_school)
          break # break out of the loop

    if not student['team_assigned']: # current student still does not have a team
      for curr_team in teams:
        # if the current team still has an empty slot
        if curr_team['current_size'] < team_size:
          update_data(curr_team, student, student_gender, student_cgpa, student_school)
          break # break out of the loop

    if not student['team_assigned']: # current student still does not have a team
      for i in range(number_of_teams):
        if (current_student_count - 1) % number_of_teams == i: # leftover students just assign to teams in ascending order
          update_data(teams[i], student, student_gender, student_cgpa, student_school)
          break

  if tutorial_group_number not in answer:
    answer[tutorial_group_number] = []
  for curr_team in teams:
    answer[tutorial_group_number].append(curr_team)
```

##**Data Visualisation**
To verify that our program has successfully distributed students across each team evenly, we created three bar graphs to illustrate the distribution based on the following prerequisites:

1) Gender Distribution

2) Average CGPA

3) School Distribution






```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# prompt user for input - tutorial group number
while True:
  try:
    tutorial_group = int(input("Enter tutorial group number (1-120): ")) # ask user for input of team_size
    if 1 <= tutorial_group <= 120:
      print(f"Displaying tutorial group G-{tutorial_group}")
      break
    else:
      print(f"{tutorial_group} is an invalid input. Please enter a tutorial group number (1-120) again")
  except ValueError: # if the user did not input an integer
    print("Invalid input. Please enter a tutorial group number (1-120) again")

# function to display Gender
def display_gender(tutorial_group, answer):
  tutorial_group_key = f"G-{tutorial_group}"

  # empty list to store plotting data
  teams = []
  male_count = []
  female_count = []

  # for each team in the specified tutorial group
  for team in answer[tutorial_group_key]:
    team_number = team["team"]
    teams.append(f"Team {team_number}")

    # count gender
    gender_count = Counter(student["gender"] for student in team["team_members"])

    # store count
    male_count.append(gender_count.get("Male", 0))
    female_count.append(gender_count.get("Female", 0))

  # display bar graph
  x = np.arange(len(teams))
  width = 0.35

  plt.figure(figsize=(10,6))

  plt.bar(x, male_count, label='Male', color='skyblue') # plot male
  plt.bar(x, female_count, bottom=male_count, label='Female', color='pink') # plot female

  plt.xlabel('Team Assigned')
  plt.ylabel('Number of Students')
  plt.title(f'Gender Distribution in Tutorial Group {tutorial_group_key}')
  plt.xticks(x, teams)
  plt.grid(True, alpha=0.3)
  plt.legend()

  plt.show()

display_gender(tutorial_group, answer)

# function to display CGPA
def display_CGPA(tutorial_group, answer):
  tutorial_group_key = f"G-{tutorial_group}"

  # empty list to store plotting data
  teams = []
  CGPA_average = []

  # for each team in the specified tutorial group
  for team in answer[tutorial_group_key]:
    team_number = team["team"]
    teams.append(f"Team {team_number}")

    # count total CGPA
    team_size = len(team["team_members"])

    CGPA_count = (sum(float(student["cgpa"]) for student in team["team_members"]) / team_size)

    # store count
    CGPA_average.append(CGPA_count)

  # display bar graph
  x = np.arange(len(teams))

  plt.figure(figsize=(10,6))

  plt.bar(x, CGPA_average, color='skyblue') # plot CGPA

  plt.xlabel('Team Assigned')
  plt.ylabel('Average CGPA')
  plt.title(f'CGPA Distribution in Tutorial Group {tutorial_group_key}')
  plt.xticks(x, teams)
  plt.grid(True, alpha=0.3)

  plt.show()

display_CGPA(tutorial_group, answer)


# function to display school
def display_school(tutorial_group, answer):
  tutorial_group_key = f"G-{tutorial_group}"

  # empty list and dictionary to store plotting data
  teams = []
  school_data = {}

  schools = [
    'CCDS', 'EEE', 'CoB (NBS)', 'SoH', 'WKW SCI', 'CoE', 'MAE',
    'SPMS', 'SBS', 'SSS', 'ASE', 'NIE', 'ADM', 'CCEB', 'MSE',
    'LKCMedicine', 'CEE', 'HASS'
    ]

  # for each team in the specified tutorial group
  for team in answer[tutorial_group_key]:
    team_number = team["team"]
    teams.append(f"Team {team_number}")

    # Count schools per team
    team_schools = Counter(student["school"] for student in team["team_members"])

    # Initialize all schools with 0 count
    school_data[team_number] = {school: 0 for school in schools}

    # store count
    for school, count in team_schools.items():
      if school in schools:  # Only count schools in our list
        school_data[team_number][school] = count

  # display bar graph
  x = np.arange(len(teams))

  plt.figure(figsize=(10,6))

  # Colours for diff schools
  colors = [
    "#8E7C93", # Muted Lavender
    "#A3A847", # Olive Green
    "#B0A4E3", # Soft Lilac
    "#E1B07E", # Soft Peach
    "#91A6C4", # Dusty Blue
    "#C0C78F", # Sage Green
    "#D4A5A5", # Rose
    "#A5B8A3", # Muted Green
    "#B6B6A5", # Grayish Beige
    "#A87C7C", # Muted Mauve
    "#C7A9A3", # Blush Beige
    "#9A9EAB", # Slate Gray
    "#BDC4A7", # Light Olive
    "#A79C93", # Taupe
    "#B5C0C4", # Soft Gray-Blue
    "#C6A49A", # Light Terracotta
    "#7C8A9A", # Steel Blue
    "#D1BFB6"  # Dusty Pink
  ]

  bottoms = np.zeros(len(teams))

  for school_idx, school in enumerate(schools):
    school_counts = [school_data[team["team"]][school] for team in answer[tutorial_group_key]]
    if any(school_counts):  # Only show in legend if school exists
        plt.bar(x, school_counts, bottom=bottoms, label=school, color=colors[school_idx])

        # Update bottoms for next school
        bottoms += np.array(school_counts)

  plt.xlabel('Team Assigned')
  plt.ylabel('Number of Students')
  plt.title(f'School Distribution in Tutorial Group {tutorial_group_key}')
  plt.xticks(x, teams)
  plt.grid(True, alpha=0.3)
  plt.legend(bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=6)

  plt.show()

display_school(tutorial_group, answer)
```
